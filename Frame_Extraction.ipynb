```python
import os, glob, cv2

PROJECT = os.getcwd()
VIDEOS_DIR = os.path.join(PROJECT, "videos")
SEL_DIR    = os.path.join(PROJECT, "frames_selected")
os.makedirs(SEL_DIR, exist_ok=True)

video_files = [
    p for p in glob.glob(os.path.join(VIDEOS_DIR, "**", "*.*"), recursive=True)
    if os.path.splitext(p)[1].lower() in [".mp4", ".mov", ".avi", ".m4v"]
]
print("Videos found:", len(video_files))
```

```python
N = 10  # every 10th frame

def safe_tag(path):
    folder = os.path.basename(os.path.dirname(path))
    name = os.path.splitext(os.path.basename(path))[0]
    return f"{folder}{name}".replace(" ", "")

total_saved = 0

for vid in sorted(video_files):
    tag = safe_tag(vid)
    out_dir = os.path.join(SEL_DIR, tag)
    os.makedirs(out_dir, exist_ok=True)

    cap = cv2.VideoCapture(vid)
    if not cap.isOpened():
        print("Cannot open:", vid)
        continue

    frame_id, saved = 0, 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_id % N == 0:
            cv2.imwrite(os.path.join(out_dir, f"frame_{frame_id:05d}.png"), frame)
            saved += 1
            total_saved += 1
        frame_id += 1

    cap.release()
    print(f" {tag}: saved {saved} frames")

print("TOTAL saved selected frames:", total_saved)
```
