```python
import os, glob
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("All imports OK")
print("OpenCV version:", cv2.__version__)
print("Current folder:", os.getcwd())
```

```python
PROJECT = os.getcwd()

VIDEOS_DIR = os.path.join(PROJECT, "videos")
SEL_DIR    = os.path.join(PROJECT, "frames_selected")
ENH_DIR    = os.path.join(PROJECT, "frames_enhanced")
FIG_DIR    = os.path.join(PROJECT, "figures")
TAB_DIR    = os.path.join(PROJECT, "tables")

for d in [SEL_DIR, ENH_DIR, FIG_DIR, TAB_DIR]:
    os.makedirs(d, exist_ok=True)

print("Videos folder:", VIDEOS_DIR)
```

```python
video_files = [
    p for p in glob.glob(os.path.join(VIDEOS_DIR, "**", "*.*"), recursive=True)
    if os.path.splitext(p)[1].lower() in [".mp4", ".mov", ".avi", ".m4v"]
]

print("Total videos found:", len(video_files))
video_files[:5]
```
